**Q0.**
1. What is the difference between regression and classification?
- Regression is using features or covariates to predict a numeric variable, while classification is the same except it's to predict a categorical variable
2. What is a confusion table? What does it help us understand about a model's performance?
- A confusion table cross tabulates the actual values and the predicted values in order to show when the model makes correct and incorrect predictions.
3. What does the SSE quantify about a particular model?
- SSE gives us a measure of the squared error generated by a model, which is how off the model is from being right. "We take the difference between true and predicted values, square it to ensure the values are positive, then sum over all predictions."
4. What are overfitting and underfitting?
- Overfitting is when a model has too much complexity for the data available and so the excess degrees of freedom are used to fit features unlikely to generalize. Underfitting is when a model is too simple for the data available and can't capture important features needed to help predict.
5. Why does splitting the data into training and testing sets, and choosing $k$ by evaluating accuracy or SSE on the test set, improve model performance?
- If we don't split the model then we won't be able to accurately test how the model will perform in practice, especially if it's an overfitting model. The model might work well on the data it was trained on because it was tuned to that specific data, but it might not work as well on similar but different data.
6. With classification, we can report a class label as a prediction or a probability distribution over class labels. Please explain the strengths and weaknesses of each approach.
- Class labels are easy to intepret but if the probabilities aren't known, then we won't be able to know to what degree of confidence the prediction was made. Probability distribution provides info about the confidence of the model's prediction, but it is harder for users to interpret probability over a simple label.

For each value of $k$, compute the mean squared error and print a scatterplot showing the test value plotted against the predicted value. What patterns do you notice as you increase $k$?
5. Determine the optimal $k$ for these data.
6. Describe what happened in the plots of predicted versus actual prices as $k$ varied, taking your answer into part 6 into account. (Hint: Use the words "underfitting" and "overfitting".)

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsRegressor
import matplotlib.pyplot as plt
```

```python
# 1
df = pd.read_csv('./data/USA_cars_datasets.csv')
df = df.loc[:,['price','year','mileage'] ]
df.head()
print(df.shape)
print(df.describe())
# No NaNs as there are 2499 rows in the dataset and 2499 observations of each variable
```

```python
# 2
def maxmin(x):
    u = (x-min(x)) / (max(x)-min(x))
    return u

df['year'] = maxmin(df['year'])
df['mileage'] = maxmin(df['mileage'])
```

```python
# 3
y = df['price']
x = df.drop('price',axis=1)
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=.2, random_state=100)
```

```python
# 4
for k in [3, 10, 25, 50, 100, 300]:
    model = KNeighborsRegressor(n_neighbors=k).fit(x_train, y_train)
    y_hat = model.predict(x_test)
    SSE = np.sum( (y_test-y_hat)**2 )
    plot, axes = plt.subplots()
    plt.scatter(y_test,y_hat)
    plt.title(f"k={str(k)}, SSE={str(SSE)}")
    axes.set_ylim(-1500, 63000)
    axes.set_xlim(-1500, 63000)
    plt.show()
# As k increases, the range of predicted values flattens, and there are less varied predictions
```

```python
# 5
k_bar = 200
k_grid = np.arange(1,k_bar)
SSE = np.zeros(k_bar)

for k in range(k_bar):
    fitted_model = KNeighborsRegressor(n_neighbors=k+1).fit(x_train, y_train)
    y_hat = fitted_model.predict(x_test)
    SSE[k] = np.sum( (y_test-y_hat)**2 )

min_index = np.argmin(SSE)
k_star = k_grid[min_index]
print(k_star)

plt.plot(np.arange(0,k_bar), SSE)
plt.xlabel("k")
plt.title(f"optimal k={str(k_star)}")
plt.ylabel('SSE')
plt.show()
```

For k values around 77, the model performs optimally, but at k = 300, the model is most likely overfitting, while for k = 3, 10, or 25, it's underfitting. When k = 50 or 100, the predictions are pretty accurate. With k = 300, there's significant horizontal clustering, where common outcomes influence predictions and grouping them unnaturally due to overfitting even though the data is evenly distributed. For small k values like 3, predictions exhibit high variance and low precision. Also, as k increases, the range of predictions narrows and outlier values become less impactful, and predictions shift closer to average.
